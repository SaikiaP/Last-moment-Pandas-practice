```python
import pandas as pd

# Creating a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Bob'],
    'Age': [28, 34, 22, None],
    'Country': ['USA', 'Canada', 'UK', 'Canada']
}
df = pd.DataFrame(data)

# Drop a column
df_dropped_column = df.drop('Age', axis=1)
print("After dropping 'Age' column:")
print(df_dropped_column)
print()

# Drop a row
df_dropped_row = df.drop(1, axis=0)
print("After dropping row with index 1:")
print(df_dropped_row)
print()

# Fill missing values with a specific value (e.g., 0)
df_filled = df.fillna(0)
print("After filling missing values:")
print(df_filled)
print()

# Replace a specific value with another value
df_replaced = df.replace({'Country': {'Canada': 'CA'}})
print("After replacing 'Canada' with 'CA' in 'Country' column:")
print(df_replaced)
print()

# Checking for duplicates based on all columns
duplicates = df.duplicated()
print("Duplicate rows:")
print(df[duplicates])
print()

# Dropping duplicates based on all columns
df_no_duplicates = df.drop_duplicates()
print("DataFrame after dropping duplicates:")
print(df_no_duplicates)
print()

# Renaming a column
df_renamed = df.rename(columns={'Age': 'Years'})
print("DataFrame after renaming 'Age' column to 'Years':")
print(df_renamed)
print()

# Sorting the DataFrame by 'Age' in ascending order
df_sorted = df.sort_values(by='Age', ascending=True)
print("DataFrame after sorting by 'Age':")
print(df_sorted)
print()

# Grouping the DataFrame by 'Country' and calculating the mean of 'Age' in each group
grouped_df = df.groupby('Country')['Age'].mean()
print("Mean age in each country:")
print(grouped_df)
```

## Data Agrregation and analysis

```python
import pandas as pd

# Creating a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [28, 34, 22, 45, 31],
    'Country': ['USA', 'Canada', 'UK', 'Australia', 'Germany'],
    'Gender': ['Female', 'Male', 'Male', 'Male', 'Female']
}
df = pd.DataFrame(data)

# Calculating the sum of 'Age' column
age_sum = df['Age'].sum()
print("Sum of 'Age' column:", age_sum)
print()

# Calculating the mean of 'Age' column
age_mean = df['Age'].mean()
print("Mean of 'Age' column:", age_mean)
print()

# Calculating the median of 'Age' column
age_median = df['Age'].median()
print("Median of 'Age' column:", age_median)
print()

# Counting the number of non-missing values in each column
column_count = df.count()
print("Number of non-missing values in each column:")
print(column_count)
print()

# Getting unique values in the 'Country' column
unique_countries = df['Country'].unique()
print("Unique countries:", unique_countries)
print()

# Counting the number of unique values in the 'Country' column
num_unique_countries = df['Country'].nunique()
print("Number of unique countries:", num_unique_countries)
print()

# Counting the occurrences of each value in the 'Gender' column
gender_counts = df['Gender'].value_counts()
print("Value counts for 'Gender' column:")
print(gender_counts)
```

```python
import pandas as pd

# Example data for time series
data = {
    'Date': ['2023-08-01', '2023-08-02', '2023-08-03', '2023-08-04', '2023-08-05'],
    'Value': [10, 15, 12, 8, 20]
}
df = pd.DataFrame(data)

# Convert the 'Date' column to datetime type
df['Date'] = pd.to_datetime(df['Date'])
print("DataFrame with datetime type 'Date' column:")
print(df)
print()

# Resample the DataFrame to a weekly frequency
df_resampled = df.resample('W', on='Date').sum()
print("DataFrame resampled to weekly frequency:")
print(df_resampled)
print()

# Shift the 'Value' column by one position
df['Value_Shifted'] = df['Value'].shift(1)
print("DataFrame with 'Value' column shifted by one position:")
print(df)
print()

# Calculate the rolling average of the 'Value' column using a window of 2 days
df['Rolling_Average'] = df['Value'].rolling(window=2).mean()
print("DataFrame with rolling average of 'Value' column:")
print(df)
```

```python
#In this example, we resampled the daily data to a weekly frequency and calculated the sum of the 'Value' column for each week. The resulting DataFrame df_weekly_sum contains the aggregated data with weekly intervals.

import pandas as pd

# Creating a sample DataFrame with daily data
data = {
    'Date': pd.date_range(start='2023-08-01', periods=10, freq='D'),
    'Value': [10, 15, 12, 8, 20, 18, 25, 30, 22, 19]
}
df = pd.DataFrame(data)

# Set 'Date' as the index
df.set_index('Date', inplace=True)

# Resample the DataFrame to weekly frequency and calculate the sum of 'Value' for each week
df_weekly_sum = df.resample('W').sum()
print(df_weekly_sum)

```

## Handling categorical data

```python
import pandas as pd

# Sample data
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [28, 34, 22, 45, 31],
    'Gender': ['Female', 'Male', 'Male', 'Male', 'Female']
}
df = pd.DataFrame(data)

# Convert the 'Age' column to float data type
df['Age'] = df['Age'].astype(float)
print("DataFrame with 'Age' column converted to float:")
print(df.dtypes)
print()

# Convert the 'Gender' column to categorical data type
df['Gender'] = df['Gender'].astype('category')
print("DataFrame with 'Gender' column converted to categorical:")
print(df.dtypes)
print()

# Get the categories of the 'Gender' column
gender_categories = df['Gender'].cat.categories
print("Categories of 'Gender' column:")
print(gender_categories)
```

## Reshaping data

```python
import pandas as pd

# Sample data for the pivot, melt, stack, and unstack examples
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Subject': ['Math', 'Math', 'Science', 'Science', 'Math'],
    'Score': [85, 78, 92, 88, 80]
}
df = pd.DataFrame(data)

# Using df.pivot to create a pivot table
pivot_table = df.pivot(index='Name', columns='Subject', values='Score')
print("Pivot Table:")
print(pivot_table)
print()

# Using df.melt to melt the DataFrame
melted_df = df.melt(id_vars=['Name', 'Subject'], value_vars=['Score'], var_name='Metric', value_name='Score_Melted')
print("Melted DataFrame:")
print(melted_df)
print()

# Using df.stack to stack the DataFrame
stacked_df = df.set_index(['Name', 'Subject']).stack().reset_index(name='Score_Stacked')
print("Stacked DataFrame:")
print(stacked_df)
print()

# Using df.unstack to unstack the DataFrame
unstacked_df = stacked_df.set_index(['Name', 'Subject', stacked_df.groupby(['Name', 'Subject']).cumcount()]).unstack().reset_index()
unstacked_df.columns = ['Name', 'Subject', 'Metric', 'Score_Unstacked']
print("Unstacked DataFrame:")
print(unstacked_df)
```

## Working with text data

```python
import pandas as pd

# Sample data
data = {
    'Name': ['Alice', ' Bob ', 'Charlie', 'David', 'eva'],
    'City': ['New York', 'London', 'Paris', 'Tokyo', 'Sydney'],
    'Phone': ['+1 123-456-7890', '+44 987-654-3210', '+33 123-456-789', '+81 987-654-321', '']
}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
print()

# Using str.lower to convert Name and City columns to lowercase
df['Name'] = df['Name'].str.lower()
df['City'] = df['City'].str.lower()
print("DataFrame after converting Name and City columns to lowercase:")
print(df)
print()

# Using str.upper to convert Phone column to uppercase
df['Phone'] = df['Phone'].str.upper()
print("DataFrame after converting Phone column to uppercase:")
print(df)
print()

# Using str.strip to remove leading and trailing whitespace from Name column
df['Name'] = df['Name'].str.strip()
print("DataFrame after removing leading and trailing whitespace from Name column:")
print(df)
print()

# Using str.split to split City column values by space and extract the first part (country)
df['Country'] = df['City'].str.split().str[0]
print("DataFrame after splitting City column values and extracting the first part (country):")
print(df)
print()

# Using str.replace to remove non-numeric characters from Phone column
df['Phone'] = df['Phone'].str.replace(r'\D', '', regex=True)
print("DataFrame after removing non-numeric characters from Phone column:")
print(df)
print()

# Using str.contains to create a new column indicating if the Name contains 'e'
df['Contains_E'] = df['Name'].str.contains('e', case=False)
print("DataFrame after creating a new column to check if the Name contains 'e':")
print(df)
```

## working with datetime data

```python
import pandas as pd

# Using pd.Timestamp to create a timestamp
timestamp = pd.Timestamp('2023-08-07 12:30:45')
print("Timestamp:")
print(timestamp)
print()

# Using pd.to_datetime to convert a list of strings to datetime objects
date_strings = ['2023-08-07', '2023-08-08', '2023-08-09']
datetime_objects = pd.to_datetime(date_strings)
print("Datetime Objects:")
print(datetime_objects)
print()

# Using Series.dt to extract components from datetime objects in a Series
data = {
    'Dates': pd.to_datetime(['2023-08-07', '2023-08-08', '2023-08-09']),
    'Values': [10, 20, 30]
}
df = pd.DataFrame(data)
df['Year'] = df['Dates'].dt.year
df['Month'] = df['Dates'].dt.month
df['Day'] = df['Dates'].dt.day
print("DataFrame with Date Components:")
print(df)
```

## Data aggregation and grouping

```python
import pandas as pd

# Sample data
data = {
    'Name': ['Alice', 'Bob', 'Alice', 'David', 'Bob', 'Alice'],
    'Subject': ['Math', 'Math', 'Science', 'Science', 'Science', 'Math'],
    'Score': [85, 78, 92, 88, 80, 95]
}
df = pd.DataFrame(data)

# Using df.groupby to group the DataFrame by 'Name'
grouped = df.groupby('Name')

# Using groupby.sum to calculate the total score for each person
total_score = grouped['Score'].sum()
print("Total Score for each person:")
print(total_score)
print()

# Using groupby.mean to calculate the average score for each person
average_score = grouped['Score'].mean()
print("Average Score for each person:")
print(average_score)
print()

# Using groupby.size to get the number of occurrences of each person
occurrences = grouped.size()
print("Number of occurrences for each person:")
print(occurrences)
print()

# Using groupby.agg to apply multiple aggregation functions at once
aggregated_data = grouped.agg({'Score': ['sum', 'mean', 'size']})
print("Aggregated Data:")
print(aggregated_data)
```
